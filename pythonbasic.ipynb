```python
a= float(input("Enter the first number :"))
b= float(input("Enter the second number :"))

choice= int(input("Enter your choice :"))

if(choice==1):
  print(a+b)
elif(choice==2):
  print(a-b)
elif(choice==3):
  print(a*b)
elif(choice==4):
  print(a/b)
else:
  print("Invalid choice")
```
