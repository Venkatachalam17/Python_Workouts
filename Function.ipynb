#Function Arguments
Default arguments

Keyword arguments

Required argument

Variable-length arguments

```python
#Default arguments
def sum(n1,n2=67):
  return n1+n2

print("The sum of numbers without default arg: ",sum(20))
print("The sum of numbers without default arg: ",sum(20,30))
```

```python
#Keyword arguments
def sum(n1,n2):
  return n1-n2

print("The sum of numbers without default arg: ",sum(20,45))
print("The sum of numbers without default arg: ",sum(n2=30,n1=20))
```

```python
#variable length argument
def function( *args_list ):   # Stores it as a Tuple
  ans = []
  for l in args_list:
    ans.append( l.upper() )                                 #This is used to convert any set of length to a tuple
  return ans
# Passing args arguments
object = function('Python', 'Functions', 'tutorial')

print( object )
```

```python
#This is used to collect length in key-value pair

def function(**kargs_list):   #Stores it as a Dictionary
    ans = []
    for key, value in kargs_list.items():
        ans.append([key, value])

    return ans

# Passing kwargs arguments
object = function(
    First="Python",
    Second="Functions",
    Third="Tutorial"
)

print(object)
```

```python
a = lambda x, y : (x * y)
print(a(4, 5))

a = lambda x, y : (x * y)
print(a(4, 5))

a = lambda x, y, z : (x + y + z)
print(a(4, 5, 5))
```

#Recursion Function
A recursive function is a function that calls itself to solve a problem by breaking it into smaller sub-problems until a base condition is reached.

```python
#function

def fact(num):
  if num == 1:
    return 1
  else:
    return (num * fact(num-1))

#Body of the program

num = int(input("Enter a number: "))
result = fact(num)
print("The factorial of the given number", num, "is", result)
```

#3. Write a Python function that takes a string as input and returns the number of vowels (a,e, i, o, u) present in the string.
* The function should be case-insensitive
* Do not count consonants or special characters

```python
def vowels(letter):
  count =0
  low = letter.lower()
  for i in low:
    if i in "aeiou":
      count+=1
  return count

print(vowels("Hello World! Python Programming."))
```
