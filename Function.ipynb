```python
#Default arguments
def sum(n1,n2=67):
  return n1+n2

print("The sum of numbers without default arg: ",sum(20))
print("The sum of numbers without default arg: ",sum(20,30))
```

```python
#Keyword arguments
def sum(n1,n2):
  return n1-n2

print("The sum of numbers without default arg: ",sum(20,45))
print("The sum of numbers without default arg: ",sum(n2=30,n1=20))
```

```python
#variable length argument
def function( *args_list ):   # Stores it as a Tuple
  ans = []
  for l in args_list:
    ans.append( l.upper() )                                 #This is used to convert any set of length to a tuple
  return ans
# Passing args arguments
object = function('Python', 'Functions', 'tutorial')

print( object )
```

```python
#This is used to collect length in key-value pair

def function(**kargs_list):   #Stores it as a Dictionary
    ans = []
    for key, value in kargs_list.items():
        ans.append([key, value])

    return ans

# Passing kwargs arguments
object = function(
    First="Python",
    Second="Functions",
    Third="Tutorial"
)

print(object)
```

#Recursion Function
A recursive function is a function that calls itself to solve a problem by breaking it into smaller sub-problems until a base condition is reached.

```python
#function

def fact(num):
  if num == 1:
    return 1
  else:
    return (num * fact(num-1))

#Body of the program

num = int(input("Enter a number: "))
result = fact(num)
print("The factorial of the given number", num, "is", result)
```

#3. Write a Python function that takes a string as input and returns the number of vowels (a,e, i, o, u) present in the string.
* The function should be case-insensitive
* Do not count consonants or special characters

```python
def vowels(letter):
  count =0
  low = letter.lower()
  for i in low:
    if i in "aeiou":
      count+=1
  return count

print(vowels("Hello World! Python Programming."))
```

#Write a Python program to perform the following tasks:

* Write a function calculate_average(marks) that returns the average of a list of marks.

* Use a lambda function to determine whether a student has passed:

  * Average ≥ 50 → Pass

  * Average < 50 → Fail

* Create a new dictionary that contains the following details for each student:

    * Student Name
    * Average Marks (rounded to 2 decimal places)
   * Result (Pass / Fail)

  * Print the details of only the students who have passed.

* Find and print the topper’s name (the student with the highest average marks).

```python
# Student marks dictionary
students = {
    "Anu": [78, 85, 90],
    "Bala": [60, 72, 68],
    "Charan": [88, 91, 84],
    "Divya": [45, 50, 40]
}

# Function to calculate average marks
def calculate_average(marks):
    return sum(marks) / len(marks)

# Lambda function to check pass or fail
result_check = lambda avg: "Pass" if avg >= 50 else "Fail"

# New dictionary to store student details
student_results = {}

# Calculate average and result for each student
for name, marks in students.items():
    avg = round(calculate_average(marks), 2)
    result = result_check(avg)
    student_results[name] = {
        "Average": avg,
        "Result": result
    }

# Print details of passed students
print("Passed Students:")
for name, details in student_results.items():
    if details["Result"] == "Pass":
        print(f"{name} - Average: {details['Average']} - Result: {details['Result']}")

# Find the topper (highest average)
topper = max(student_results, key=lambda x: student_results[x]["Average"])
print("\nTopper:", topper)
```

# Write a recursive Python program to analyze a given positive integer.
You must NOT use loops (for, while).
Given an integer n, write recursive functions to:
* Count the number of digits
* Find the sum of digits
* Find the maximum digit
* Reverse the number

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)

def sum_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_digits(n // 10)

def max_digit(n):
    if n < 10:
        return n
    m = max_digit(n // 10)
    return n % 10 if n % 10 > m else m

def reverse_num(n, r=0):
    if n == 0:
        return r
    return reverse_num(n // 10, r * 10 + n % 10)

n = int(input("Enter number: "))

print("Digits:", count_digits(n))
print("Sum:", sum_digits(n))
print("Max Digit:", max_digit(n))
print("Reverse:", reverse_num(n))
```
