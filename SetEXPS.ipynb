```python
def noofvowels(text):
  vowels={"a","e","i","o","u"}
  count=0
  for i in text.lower():
    if i in vowels:
      count+=1
  return count
print(noofvowels("woRld"))
```

```python
students = {
    "Anu": [78, 85, 90],
    "Bala": [60, 72, 68],
    "Charan": [88, 91, 84],
    "Divya": [45, 50, 40]
}
def calculate_average(marks):
    return sum(marks) / len(marks)
check_result = lambda avg: "Pass" if avg >= 50 else "Fail"
student_results = {}
for name, marks in students.items():
    avg = round(calculate_average(marks), 2)
    result = check_result(avg)
    student_results[name] = {
        "Average": avg,
        "Result": result
    }
print("Passed Students:")
for name, details in student_results.items():
    if details["Result"] == "Pass":
        print(f"{name} - Average: {details['Average']}")
topper = max(student_results, key=lambda x: student_results[x]["Average"])
print("\nTopper:", topper)
```

```python
def counts(n):
    if n == 0:
        return 0
    return 1 + counts(n // 10)
def sums(n):
    if n == 0:
        return 0
    return (n % 10) + sums(n // 10)
def maxi(n):
    if n < 10:
        return n
    return max(n % 10, maxi(n // 10))
def reverses(n, rev=0):
    if n == 0:
        return rev
    return reverses(n // 10, rev * 10 + n % 10)
num = 53829
print("No of digits:",counts(num))
print("Sum:",sums(num))
print("Maximum:",maxi(num))
print("Reversed:",reverses(num))
```
