#1.A university library management system stores book details in a Python dictionary named books, where the is the book_id and the is another dictionary containing title, author, andavailable_copies.

During end-of-day processing, the librarian performs the following operations:

* Creates a backup of the current books record before making changes.

* Initializes a new dictionary of default categories ("Science", "Arts", "Engineering")
with an initial count of 0.

* Safely retrieves the number of available copies for a given book_id without causing
an error if the ID does not exist.

* Displays all book IDs with their corresponding details for reporting.

* Extracts only the list of book IDs and only the list of available copy counts for
analytics.


* Removes the most recently added book entry during a data correction step.

* Ensures a missing book category is added with a default value if it does not already
exist.

* Merges a newly received shipment of books (stored in another dictionary) into the
existing books dictionary.

* Finally, clears all records at the end of the academic year.

* Identify and justify the appropriate Python dictionary methods (clear(), copy(),
fromkeys(), get(), items(), keys(), pop(), popitem(), setdefault(), values(), update())
that should be used for each of the above operations, and explain each method is
suitable in the given scenario.

```python
import builtins

# Initial books dictionary
books = {
    101: {"title": "Python Basics", "author": "Guido", "available_copies": 5},
    102: {"title": "Data Structures", "author": "Mark", "available_copies": 3},
    103: {"title": "AI Concepts", "author": "Andrew", "available_copies": 7}
}

# 1. Create a backup of the books record
books_backup = books.copy()

# 2. Initialize default categories with initial count 0
categories = dict.fromkeys(["Science", "Arts", "Engineering"], 0)

# 3. Safely retrieve available copies for a given book_id
book_id_to_check = 104
copies = books.get(book_id_to_check, {}).get("available_copies", 0)
print(f"Available copies for book_id {book_id_to_check}: {copies}")

# 4. Display all book IDs with their details
print("\nAll books and details:")
for book_id, details in books.items():
    print(book_id, details)

# 5. Extract only the list of book IDs
book_ids = builtins.list(books.keys())
print("\nBook IDs:", book_ids)

# 6. Extract only the list of available copy counts
available_copies_list = [details["available_copies"] for details in books.values()]
print("Available copies list:", available_copies_list)

# 7. Remove a book record when permanently discarded
removed_book = books.pop(102, None)
print("\nRemoved book with ID 102:", removed_book)

# 8. Remove the most recently added book entry
last_removed = books.popitem()
print("Most recently removed book:", last_removed)

# 9. Ensure a missing category is added with a default value
categories.setdefault("Medical", 0)
print("\nCategories after setdefault:", categories)

# 10. Merge newly received shipment of books
new_shipment = {
    104: {"title": "Machine Learning", "author": "Tom", "available_copies": 6},
    105: {"title": "Cloud Computing", "author": "Sam", "available_copies": 4}
}
books.update(new_shipment)
print("\nBooks after new shipment:")
for book_id, details in books.items():
    print(book_id, details)

# 11. Clear all records at the end of the academic year
books.clear()
print("\nBooks after clearing:", books)
```
